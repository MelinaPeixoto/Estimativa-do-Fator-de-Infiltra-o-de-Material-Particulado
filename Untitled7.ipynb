````python
import requests
import pandas as pd
import io
import re # Importando a biblioteca de expressões regulares

ckan_url = "https://opendatasus.saude.gov.br/api/3/action/package_show?id=sim"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    pkg_response = requests.get(ckan_url, headers=headers, timeout=10)
    print(f"HTTP Status Code for package_show: {pkg_response.status_code}")
    pkg_response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    pkg = pkg_response.json()
except requests.exceptions.RequestException as e:
    print(f"Error fetching package info from {ckan_url}: {e}")
    if pkg_response is not None:
        print(f"Status Code: {pkg_response.status_code}")
        print(f"Response content (first 500 chars):\n{pkg_response.text[:500]}")
    pkg = None

if pkg:
    resources = pkg['result']['resources']
    all_csv_links = [r['url'] for r in resources if r['format'] == 'CSV']

    # Filtra os links para o período de 2015 a 2025
    filtered_csv_links = []
    start_year = 2015
    end_year = 2025

    for link in all_csv_links:
        # Extrai o ano do nome do arquivo (ex: Mortalidade_Geral_1995.csv)
        match = re.search(r'(\d{4})\.csv$', link)
        if match:
            year = int(match.group(1))
            if start_year <= year <= end_year:
                filtered_csv_links.append(link)

    csv_links = filtered_csv_links
    print(f"Filtrados {len(csv_links)} links CSV para os anos de {start_year}-{end_year}.")

    dfs = []
    for link in csv_links:
        try:
            print(f"Attempting to download from: {link}")
            file_response = requests.get(link, headers=headers, timeout=60) # Increased timeout for large files
            print(f"HTTP Status Code for file download: {file_response.status_code}")
            file_response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
            # Added low_memory=False to potentially handle DtypeWarnings better
            df = pd.read_csv(io.StringIO(file_response.content.decode('latin1')), sep=';', low_memory=False)
            dfs.append(df)
            print("OK:", link)
        except requests.exceptions.RequestException as e:
            print(f"Erro ao baixar {link}: {e}")
            if file_response is not None:
                print(f"Status Code: {file_response.status_code}")
                print(f"Response content (first 500 chars):\n{file_response.text[:500]}")

    if dfs:
        df_sim = pd.concat(dfs)
        print(f"df_sim successfully created with shape: {df_sim.shape}") # Confirmation message
        df_sim.head()
    else:
        print("No dataframes were successfully downloaded.")
else:
    print("Could not retrieve package information.")
````

#### b) Converter para string (object)
Se a coluna contém uma mistura genuína de texto e números que devem ser tratados como texto (por exemplo, códigos alfanuméricos), você pode convertê-la explicitamente para o tipo `string` (ou `object` em Pandas).


````python
# Exemplo: Converter a coluna 23 (índice 22) para string
# Substitua 'NOME_DA_SUA_COLUNA_STRING' pelo nome real da coluna
# Ou use df_sim.iloc[:, 22] para selecionar por índice

if df_sim.shape[1] > 22:
    col_name_string = df_sim.columns[22]
    df_sim[col_name_string] = df_sim[col_name_string].astype(str)
    print(f"Coluna '{col_name_string}' convertida para string.")
    print(f"Novos tipos de dados da coluna '{col_name_string}': {df_sim[col_name_string].dtype}")
else:
    print(f"A coluna com índice 22 não existe para conversão para string.")
````

#### c) Usar `low_memory=False` no `pd.read_csv` (se o erro for apenas de inferência)
Se o problema for apenas o Pandas tendo dificuldade em inferir o tipo de dados ao ler grandes arquivos (e não necessariamente uma mistura *real* de dados incompatíveis), você pode tentar adicionar `low_memory=False` ao `pd.read_csv`. Código pronto:

````python
def allocate_sales(df_sales, df_pam):
    area_state = df_pam.groupby(['state_code','year'])['area_total_ha'].sum().reset_index()
    area_state = area_state.rename(columns={'area_total_ha':'state_area'})

    tmp = df_pam.merge(area_state, on=['state_code','year'])
    tmp = tmp.merge(df_sales, on=['state_code','year'], how='left')

    tmp['use_estimated_kg'] = tmp['sales_kg'] * (tmp['area_total_ha'] / tmp['state_area'])

    return tmp[['codmun','year','ingredient','use_estimated_kg']]
````

5. CONSTRUÇÃO DO ÍNDICE (IIEAA)

Usa 4 variáveis principais:

área plantada municipal (log)

produção municipal (log)

composição de culturas de alto uso

uso estimado por hectare (log)

score CEM (opcional, se fizer a matriz)

Código PCA:

````python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

df['X1'] = np.log1p(df['area_total_ha'])
df['X2'] = np.log1p(df['total_prod_t'])
df['X3'] = df['share_high_use']
df['X4'] = np.log1p(df['use_estimated_kg'] / df['area_total_ha'])

X = StandardScaler().fit_transform(df[['X1','X2','X3','X4']])
df['IIEAA'] = PCA(n_components=1).fit_transform(X)
````

6. MODELO EPIDEMIOLÓGICO
Poisson com offset populacional:

````python
import statsmodels.formula.api as smf

df['offset'] = np.log(df['pop'])

model = smf.glm(
    formula='deaths_cancer ~ IIEAA + year',
    data=df,
    family=sm.families.Poisson(),
    offset=df['offset']
).fit()

print(model.summary())
````

## 4. Funções utilitárias úteis
- funcoes para alocar vendas estaduais para municípios
- funções para padronizar por idade (esboço)

````python
# 4.1 Alocar vendas estaduais para municípios por área/producao (funções)
def allocate_sales_by_area(sales_state_df, pam_df):
    """
    sales_state_df: DataFrame with columns ['state_code','year','ingredient','sales_kg']
    pam_df: DataFrame with columns ['codmun','state_code','year','area_total_ha','area_by_crop'...]
    Returns: DataFrame with columns ['codmun','year','ingredient','sales_alloc_kg']
    """
    # sum area per state-year
    area_state = pam_df.groupby(['state_code','year'])['area_total_ha'].sum().reset_index().rename(columns={'area_total_ha':'area_state'})
    pam = pam_df.merge(area_state, on=['state_code','year'], how='left')
    # merge sales
    merged = pam.merge(sales_state_df, on=['state_code','year'], how='left')
    # allocate
    merged['sales_alloc_kg'] = merged['sales_kg'] * (merged['area_total_ha'] / merged['area_state'])
    return merged[['codmun','year','ingredient','sales_alloc_kg']]

# 4.2 Função de padronização simples (direct standardization placeholder)
def compute_asr(counts_df, pop_df, standard_pop):
    """
    counts_df: DataFrame with columns ['codmun','year','age_group','deaths']
    pop_df: DataFrame with columns ['codmun','year','age_group','pop']
    standard_pop: dict or DataFrame with age_group -> standard population
    Returns ASR per codmun-year
    """
    df = counts_df.merge(pop_df, on=['codmun','year','age_group'], how='left')
    df['rate'] = df['deaths'] / df['pop']
    df['expected'] = df['rate'] * df['pop']  # trivial; actual standardization needs standard pop
    # Placeholder: return crude rates per 100k
    agg = df.groupby(['codmun','year']).apply(lambda g: g['deaths'].sum() / g['pop'].sum() * 100000).reset_index()
    agg.columns = ['codmun','year','crude_rate_per100k']
    return agg
````

Para testar o fluxo sem dados reais: geramos uma amostra sintética, construímos IIEAA via PCA, ajustamos um modelo Poisson e salvamos outputs.


````python
# 5. Criar dataset sintético para demonstração (substituir por dados reais)
rng = np.random.default_rng(42)
n = 600  # número de observações município-ano sintéticas
df = pd.DataFrame({
    'codmun': np.arange(100100,100100+n),
    'year': rng.integers(2000,2024,size=n),
    'state_code': rng.integers(11,43,size=n),  # placeholder
    'area_total_ha': np.abs(rng.normal(3000, 9000, size=n)),
    'total_prod_t': np.abs(rng.normal(15000, 40000, size=n)),
    'share_high_use': np.clip(rng.normal(0.25,0.18,size=n),0,1),
    'sales_alloc_per_ha': np.abs(rng.normal(30,150,size=n)),
    'pop': np.abs(rng.integers(800,200000,size=n))
})

# Transformações iniciais e criação de Xs para PCA
df['X1'] = np.log1p(df['area_total_ha'])
df['X2'] = np.log1p(df['total_prod_t'])
df['X3'] = df['share_high_use']
df['X4'] = np.log1p(df['sales_alloc_per_ha'])

scaler = StandardScaler()
X = scaler.fit_transform(df[['X1','X2','X3','X4']])
pca = PCA(n_components=1)
df['IIEAA_pca'] = pca.fit_transform(X)[:,0]

# Gerar contagens de óbitos sintéticas (Poisson) com offset pop
beta0 = -7.0
beta1 = 0.45
mu = np.exp(beta0 + beta1*df['IIEAA_pca'] + np.log(df['pop']/1000))
rng2 = np.random.default_rng(123)
df['deaths_cancer'] = rng2.poisson(mu)

df[['codmun','year','pop','IIEAA_pca','deaths_cancer']].head()
````

````python
# 7. Modelo Poisson (com offset). Verificar overdispersion depois.
df['offset'] = np.log(df['pop'])
formula = 'deaths_cancer ~ IIEAA_pca + year'
model = smf.glm(formula=formula, data=df, family=sm.families.Poisson(), offset=df['offset'])
res = model.fit()
print(res.summary())

# Check overdispersion: ratio deviance/df
deviance = res.deviance
df_resid = res.df_resid
print('Deviance/df:', deviance/df_resid)
````

## 6. Crop-Exposure Matrix (CEM) — scaffold (exemplo)
A CEM exige curadoria: matriz culture × ingredient com valores kg/ha ou doses/ha.
Abaixo segue um esqueleto de como aplicar a matriz ao PAM para gerar uma estimação de exposição ponderada por toxicidade.


````python
# 9. Scaffolding para CEM (exemplo)
# Exemplo de matriz simplificada: culturas x ingredientes (valores fictícios: kg/ha)
cem = pd.DataFrame({
    'ingredient': ['ingA','ingB','ingC'],
    'cult_soja': [0.8, 0.1, 0.0],
    'cult_algodao': [1.2, 0.2, 0.1],
    'cult_milho': [0.3, 0.5, 0.0]
})
cem = cem.set_index('ingredient').T.reset_index().rename(columns={'index':'culture'})
# Simular PAM por cultura por mun/ano (exemplo)
pam_cult = pd.DataFrame({
    'codmun': df['codmun'],
    'year': df['year'],
    'cult_soja_ha': np.round(df['area_total_ha'] * df['share_high_use'] * 0.6),
    'cult_algodao_ha': np.round(df['area_total_ha'] * df['share_high_use'] * 0.2),
    'cult_milho_ha': np.round(df['area_total_ha'] * (1-df['share_high_use']) * 0.5),
})

# calcular exposição por ingrediente approximada:
# Exposure = sum_cult (area_cult_ha * average_kgperha_for_cult)
# Para demo, construímos um vector média kg/ha por cultura
avg_kg_per_ha = {'cult_soja': 1.0, 'cult_algodao': 1.3, 'cult_milho': 0.6}
pam_cult['exposure_kg'] = (pam_cult['cult_soja_ha'] * avg_kg_per_ha['cult_soja'] +
                          pam_cult['cult_algodao_ha'] * avg_kg_per_ha['cult_algodao'] +
                          pam_cult['cult_milho_ha'] * avg_kg_per_ha['cult_milho'])
# merge to main df
df = df.merge(pam_cult[['codmun','year','exposure_kg']], on=['codmun','year'], how='left')
# normalize and create IIEAA_cem
df['IIEAA_cem'] = np.log1p(df['exposure_kg'])
df['IIEAA_cem_z'] = (df['IIEAA_cem'] - df['IIEAA_cem'].mean())/df['IIEAA_cem'].std()
df[['codmun','year','exposure_kg','IIEAA_cem_z']].head()
````

````python
def allocate_sales_by_area(sales_state_df, pam_df):
    """
    sales_state_df: DataFrame with columns ['state_code','year','ingredient','sales_kg']
    pam_df: DataFrame with columns ['codmun','state_code','year','area_total_ha','area_by_crop'...]
    Returns: DataFrame with columns ['codmun','year','ingredient','sales_alloc_kg']
    """
    # sum area per state-year
    area_state = pam_df.groupby(['state_code','year'])['area_total_ha'].sum().reset_index().rename(columns={'area_total_ha':'area_state'})
    pam = pam_df.merge(area_state, on=['state_code','year'], how='left')
    # merge sales
    merged = pam.merge(sales_state_df, on=['state_code','year'], how='left')
    # allocate
    merged['sales_alloc_kg'] = merged['sales_kg'] * (merged['area_total_ha'] / merged['area_state'])
    return merged[['codmun','year','ingredient','sales_alloc_kg']]


use_estimated_kg_df = allocate_sales_by_area(df_sales, df_pam_processed)

print("\nuse_estimated_kg_df head:")
display(use_estimated_kg_df.head())
print("\nuse_estimated_kg_df info:")
display(use_estimated_kg_df.info())

# Verification of total sales (optional, for debugging/validation)
print("\nVerificação da alocação de vendas:")
print(f"Total de vendas estaduais em df_sales: {df_sales['sales_kg'].sum():,.2f} kg")
print(f"Total de uso estimado municipal em use_estimated_kg_df: {use_estimated_kg_df['sales_alloc_kg'].sum():,.2f} kg")
````
